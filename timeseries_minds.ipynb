```
from mindsdb_native import Predictor
mindsdb = Predictor(name="ny_energy")
mindsdb.learn(
    from_data='https://mindsdb-web-test.s3.amazonaws.com/nyc_energy_consumption.csv', 
    to_predict='temp' 
)
```

```
result = mindsdb.predict(when_data={'demand': 500})
# print(result[0]['temp'])
print(result)
```

```
mindsdb_timeseries = Predictor(name="timeseries")
mindsdb_timeseries.learn(
    from_data='https://mindsdb-web-test.s3.amazonaws.com/timeSeries.csv', 
    to_predict='Sunspots' 
)
```

```
result_timeseries = mindsdb_timeseries.predict(when_data={'Month': '1983-12'})
# print(result[0]['temp'])
print(result_timeseries)
```
